```python
#Multi-Part Algorithms: Add vs. Multiply 

#Add the Runtimes
A_array=list(range(0,100))
B_array=list(range(0,200))

for i in A_array:
  print(i)

for j in B_array:
  print(j)
  

```

This one takes O(A+B) because two for loops in the form "do this, then, when you're all done, do that"

```python
#Multiply the runtimes

for i in A_array:
  for j in B_array:
    print(str(i)+','+str(j))
```

In the example on the right, we do B chunks of work for each element in A. Therefore, the total amount of
workisO(A * B) .

```python
given_list=[4,7,2]
def total(given_list):
  x=0
  for item in given_list:
    x+=item
  return x

total(given_list)
```
